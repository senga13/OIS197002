```r
v1_annees<- c(1:1500) 
v2_populations<-v1_annees^2
plot(v1_annees,
    v2_populations,
     xlab= "Années",
     ylab= "Augmentation au carré de la population",
     type= "p" ,
     cex.lab= 1.5,
     col= " blue" ,
    main= " Evolution de la population dans un monde imaginaire",
    lwd= 3)

legend("topleft",
       legend= "population",
       col= c("blue") ,
       lty= c(1,1) )


    
      
      
      

```
